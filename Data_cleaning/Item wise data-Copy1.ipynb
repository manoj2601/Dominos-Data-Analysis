```python
import os
import sys
import pandas as pd
import re
```

```python
path = '/media/root/data/analysis/data/Item_wise_data_Recieved/'
```

```python
cities = ["Bhopal/", "Jaipur/", "Mumbai/" , "Delhi/", "Kolkata/"]
cities
```

```python
dic = {}
for city in cities:
    l2 = os.listdir(path+city)
    for x in l2:
        m = "Dec" if 'Dec' in x else "July"
        pat = r"[-|\s]([a-zA-Z]+)-(\d+)-(\d+)[-]+([a-zA-Z]+)"
        a = re.search(pat, x)
        b = a.group(2) + '-' + a.group(3)
        c = (city[:-1], m, b)
        print(c)

        if c not in dic:
            df = pd.read_excel(path+city+x)
            dic[c] = df
        else:
            df = dic[c]
        
#         df.groupby(["Order_Date", "city"]).size().reset_index(name="Time")
        
print(len(dic))       
        
```
