```python
city = 'Bhopal'
fileName = f'edgesLatLong_{city}.csv'
```

```python
LatLongs = []
with open(fileName, 'r') as f:
    Lines = f.readlines()
    for line in Lines:
        a = line.strip().split(',')
        l = []
        for i in range(0, 4):
            l.append(float(a[i]))
        LatLongs.append(l)
```
