```python
# data read
PATH = f'/media/root/data/analysis/data/Item_wise_data/{city}'
df = pd.read_excel(f'{PATH}/Item_wise-{city}-17-23-Dec-2021.xlsx')
df = df.sort_values(by =['Order_Date', 'Location_Code', 'Order_Number'])
```

```python
df = df[pd.notna(df['Oven_Time'])]
```

```python
def get_preparation_time(order):
    diff = order['Oven_Time'] - order['Kitchen_Display_Time']
    if(diff.total_seconds() <= 0):
        print("HEYY", diff.total_seconds())
    return 420+diff.total_seconds()
```

```python
def getSlot(t, slotTime):
    return round(t/slotTime);
```

```python
sampleOrder = df.iloc[0]
sampleOrder
```

```python
dic = {}
cnt=0
prepTime = []
slotWidth = 5
for index, order in df.iterrows():
    prep_time = get_preparation_time(order)
    slot = getSlot(prep_time, slotWidth)
    if(prep_time > 30*60):
        cnt+=1
        continue
    for i in range(0, order['Quantity']):
        prepTime.append(prep_time)
    if slot in dic:
        dic[slot] += order['Quantity']
    else:
        dic[slot] = order['Quantity']
```

```python
X = np.array([i/(60//slotWidth) for i in range(6*(60//slotWidth), 21*(60//slotWidth))])
y = np.array([dic[i] if i in dic else 0 for i in range(6*(60//slotWidth), 21*(60//slotWidth))])
plt.plot(X, y)

plt.xticks([i for i in range(6, 21)])
plt.title(f"Distribution of Preparation Time (minutes)\naverage prep time = {np.mean(prepTime)/60:.2f} minutes\naverage std time = {np.std(prepTime)/60:.2f} minutes")
plt.xlabel("Preparation Time (minutes)")
plt.ylabel("Items corresponding to prep time in duration 17-23 dec 2021")
plt.grid()
plt.show()
```

```python
total_slots = 24

def getTimeIndex(dt):
	current = datetime(2011, 1, 1)
	t1 = current.time()
	t2 = dt.time()
	for i in range(0, total_slots):
		current = current + timedelta(minutes = 60)
		if(current.time() > t2):
			return i
	return total_slots-1
```

```python
allY = []
meanX = []
stdX = []

finalDict = []
for i in range(0, total_slots):
	finalDict.append([])
    
for index, order in df.iterrows():
    dt = order['Kitchen_Display_Time']
    i = getTimeIndex(dt)
    t = get_preparation_time(order)
    finalDict[i].append(t+7.0)
```

```python

X = [i for i in range(0, total_slots)]
y = []
e = []
for i in range(0, total_slots):
	if(len(finalDict[i]) == 0):
		y.append(0)
		e.append(0)
	else:
		y.append(np.mean(finalDict[i]))
		e.append(np.std(finalDict[i]))

# plt.plot(X, y)
plt.errorbar(X, y, e, fmt='-o')
plt.xticks(X)
plt.title(f"Avg food preperation time in each timeslot for 17-31 dec 2021 in {city}")
plt.xlabel('Time (in 24 hour format)')
plt.ylabel('Time in minutes')
plt.ylim(0, None)
plt.show()
```
